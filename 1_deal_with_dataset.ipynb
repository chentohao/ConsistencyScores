```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/heart-disease/UCI-Heart-Disease-Dataset.csv")
```

```python
df.shape
```

```python
df['sex'][df['sex']==0]='female'
df['sex'][df['sex']==1]='male'

df['cp'][df['cp']==1]='typical angina'
df['cp'][df['cp']==2]='atypical angina'
df['cp'][df['cp']==3]='non-anginal pain'
df['cp'][df['cp']==4]='asymptomatic'

df['fbs'][df['fbs']==0]='false'
df['fbs'][df['fbs']==1]='true'

df['restecg'][df['restecg']==0]='nomal'
df['restecg'][df['restecg']==1]='having ST-T wave abnormality'
df['restecg'][df['restecg']==2]='left ventricular hypertrophy'

df['exang'][df['exang']==0]='no'
df['exang'][df['exang']==1]='yes'

df['slope'][df['slope']==1]='upsloping'
df['slope'][df['slope']==2]='flat'
df['slope'][df['slope']==3]='downsloping'

df['thal'][df['thal']==0]='normal'
df['thal'][df['thal']==1]='fixed defect'
df['thal'][df['thal']==2]='reversable defect'
```
