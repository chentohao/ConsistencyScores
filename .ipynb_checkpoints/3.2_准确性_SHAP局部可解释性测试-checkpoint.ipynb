```python
#导入工具包、预处理后的数据集
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../code/breast-cancer-wisconsin.csv')

#划分特征列和标签列
#矩阵用大写，向量用小写
X = df.drop('diagnosis',axis=1)

y=df['diagnosis']

feature_names = X.columns

#类标整数化
from sklearn.preprocessing import LabelEncoder#该类使用0到n_classs-1之间的值对目标标签进行编码。该转换器应用于编码目标值y，而不是输入X
le = LabelEncoder()
y = le.fit_transform(y)

#划分训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#保存数据集

#保留原始数据集中的索引
#feature_test = pd.DataFrame(X_test, columns=feature_names)
#feature_test = np.copy(X_test)
#feature_test = pd.DataFrame(data=feature_test[0:,0:], columns=feature_names)
#label_test = pd.DataFrame(y_test, columns=['diagnosis'])
#test_data = 'test_data.csv'
#pd.concat([label_test,feature_test], axis=1).to_csv(test_data, index=False)

#feature_train = np.copy(X_train)
#feature_train = pd.DataFrame(data=feature_train[0:,0:], columns=feature_names)
#label_train = pd.DataFrame(y_train, columns=['diagnosis'])
#train_data = 'train_data.csv'
#pd.concat([label_train,feature_train], axis=1).to_csv(train_data, index=False)

from sklearn.preprocessing import StandardScaler
transform = StandardScaler() #实例化转换器
#标准化，保证每个维度的特征数据方差为1，均值为0，使得预测结果不会被某些维度过大的特征值而主导
X_train = transform.fit_transform(X_train)
X_test = transform.transform(X_test)

import joblib
#load joblib
model = joblib.load('../code/saved_model/model.pkl')

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
```

```python
#导入SHAP机器学习可解释性分析工具包
import shap
#初始化javascript环境，也就是绘图环境
shap.initjs()
```

```python
#计算测试集每个样本的每个特征对两类预测结果的SHAP值
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)
```

```python
temp = np.copy(X_test)
temp = pd.DataFrame(data=temp[0:,0:], columns=feature_names)
temp
```

```python
#选取测试集中索引为idx的样本
idx=170
patient=temp.iloc[idx,:]
patient_df=temp.loc[idx:idx]
model_predict_proba = model.predict_proba(patient_df)[0][1]
print('测试集中索引为{}的病人是否患恶性乳腺癌肿瘤的真实标签是{}，模型预测恶性的概率为{:.2f}'.format(idx,bool(y_test[idx]),model_predict_proba))

shap_values_patient = explainer.shap_values(patient)
shap.force_plot(explainer.expected_value[1],shap_values_patient[1],patient)
```

```python
#选取测试集中索引为idx的样本
idx=169
patient=temp.iloc[idx,:]
patient_df=temp.loc[idx:idx]
model_predict_proba = model.predict_proba(patient_df)[0][1]
print('测试集中索引为{}的病人是否患恶性乳腺癌肿瘤的真实标签是{}，模型预测恶性的概率为{:.2f}'.format(idx,bool(y_test[idx]),model_predict_proba))

shap_values_patient = explainer.shap_values(patient)
shap.force_plot(explainer.expected_value[1],shap_values_patient[1],patient)
```

```python
#测试集所有样本的summary plot
#将测试集所有样本的force plot旋转90度并拼接在一起，形成summary plot
#可以在下拉菜单选择按照相似性聚类展示、按照预测结果概率从大到小展示、按照测试集原本样本顺序、按照某个特征分别展示
number_show=60
shap_value_summary = explainer.shap_values(temp.iloc[:number_show])
shap.force_plot(explainer.expected_value[1],shap_value_summary[1],temp.iloc[:number_show])
```

```python
class ShapObject:
    
    def __init__(self, base_values, display_data, values, feature_names):
        self.base_values = base_values # Single value
        self.display_data = display_data # Raw feature values for 1 row of data
        self.values = values # SHAP values for the same row of data
        self.feature_names = feature_names # Column names
        
idx=169
patient=temp.iloc[idx,:]
shap_values_patient = explainer.shap_values(patient)
shap_object = ShapObject(base_values = explainer.expected_value[1],
                         values = shap_values_patient[1],
                         feature_names = feature_names,
                         display_data = patient)

shap.waterfall_plot(shap_object,max_display=30)
```

```python
df = pd.read_csv('explain_set/shap/local_shap/n=7/shap_local_explain.csv')
X_explain = df.drop('diagnosis',axis=1)
y_explain=df['diagnosis']

#标准化，保证每个维度的特征数据方差为1，均值为0，使得预测结果不会被某些维度过大的特征值而主导
X_explain = transform.transform(X_explain)

y_pred_explain = model.predict(X_explain)

y_pred_proba_explain = model.predict_proba(X_explain)

#混淆矩阵
from sklearn.metrics import confusion_matrix
confusion_matrix_model = confusion_matrix(y_explain,y_pred_explain)
print("Confusion Matrix:\n", confusion_matrix_model)

from sklearn.metrics import classification_report
print(classification_report(y_explain,y_pred_explain,target_names=['良性','恶性']))
```

```python
from sklearn.metrics import recall_score
print(recall_score(y_explain, y_pred_explain))
```

```python
from sklearn.metrics import roc_auc_score
print("AUC指标：", roc_auc_score(y_explain,y_pred_explain))
```
