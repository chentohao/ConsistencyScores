```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus']=False
sns.set_theme(style='darkgrid')
result = pd.read_csv('result/heart-disease/validity/validity.csv')
result
```

```python
g = sns.FacetGrid(result, col='Model', hue="Interpreter")
g.map_dataframe(sns.lineplot, x='Number of removed features', y='validity score')
g.add_legend()

plt.ylim(0, 1)
plt.show()
```

```python
g.savefig("result/heart-disease/validity/validity.svg",dpi=400)
```
