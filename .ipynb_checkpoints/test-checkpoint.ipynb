```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

df = pd.read_csv('dataset/early-stage-diabetes-risk-prediction/diabetes.csv')

X = df.drop('class',axis=1)

y=df['class']

feature_names = X.columns

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.preprocessing import StandardScaler
transform = StandardScaler()
X_train['Age'] = transform.fit_transform(X_train[['Age']])
X_test['Age'] = transform.transform(X_test[['Age']])

import joblib
model = joblib.load('saved_model/early-stage-diabetes-risk-prediction/MLP.pkl')

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
```

```python
from anchor import anchor_tabular
class_names=['Negative', 'Positive']
categorical_names = {
    1: ['Female', 'Male'], 2: ['Female', 'Male'], 3: ['No', 'Yes'], 4: ['No', 'Yes'], 5: ['No', 'Yes'], 6: ['No', 'Yes'], 7: ['No', 'Yes'], 8: ['No', 'Yes'], 9: ['No', 'Yes'], 10: ['No', 'Yes'],
    11: ['No', 'Yes'], 12: ['No', 'Yes'], 13: ['No', 'Yes'], 14: ['No', 'Yes'], 15: ['No', 'Yes'], 16: ['No', 'Yes'], 17: ['No', 'Yes'], 18: ['No', 'Yes'], 19: ['No', 'Yes'],
    20: ['No', 'Yes'], 21: ['No', 'Yes'], 22: ['No', 'Yes'], 23: ['No', 'Yes'], 24: ['No', 'Yes'], 25: ['No', 'Yes'], 26: ['No', 'Yes'], 27: ['No', 'Yes'], 28: ['No', 'Yes'], 29: ['No', 'Yes'], 30: ['No', 'Yes']
}
explainer = anchor_tabular.AnchorTabularExplainer(
    class_names,
    feature_names,
    np.array(X_train),
    categorical_names)
```

```python
explanations = [] 

# First loop over idx, generate explanation, and store the results in explanations
for idx in range(156):
    exp = explainer.explain_instance(X_test.iloc[idx].values.reshape(1, -1), model.predict,threshold=0.95)
    explanations.append(exp)
```

```python
# Then loop over k, and for each k iterate through all explanations
for k in range(1, 32):
    test_data = pd.read_csv('dataset/early-stage-diabetes-risk-prediction/test_data.csv')
    count = 0
    maxlen = 0
    for idx in range(156):
        exp = explanations[idx] # Use the previously stored explanation
        lista = exp.features()
        length = len(lista)
        count = count + length
        if length > maxlen:
            maxlen = length
        if length > k:
            for j in range(0,k):
                if lista[j]==0:
                    test_data.iat[idx,lista[j]]=94
                else:
                    test_data.iat[idx,lista[j]]=0
        elif length <= k:
            for j in range(0,length):
                if lista[j]==0:
                    test_data.iat[idx,lista[j]]=94
                else:
                    test_data.iat[idx,lista[j]]=0
    filename = 'explain_set/early-stage-diabetes-risk-prediction/MLP/anchor/anchor_explain_'+str(k)+'.csv'
    test_data.to_csv(filename,index=False)
    mean_length = count / 156
print("mean_length:",mean_length)
print("maxlen:",maxlen)
```

```python
for k in range(1, 32):
    test_data = pd.read_csv('dataset/early-stage-diabetes-risk-prediction/test_data.csv')
    keep_temp = pd.read_csv('dataset/early-stage-diabetes-risk-prediction/keep_temp.csv')
    count = 0
    maxlen = 0
    for idx in range(156):
        exp = explanations[idx] # Use the previously stored explanation
        lista = exp.features()
        length = len(lista)
        count = count + length
        if length > maxlen:
            maxlen = length
        if length > k:
            for j in range(0,k):
                keep_temp.iat[idx,lista[j]]=test_data.iat[idx,lista[j]]
        elif length <= k:
            for j in range(0,length):
                keep_temp.iat[idx,lista[j]]=test_data.iat[idx,lista[j]]
    filename = 'explain_set/early-stage-diabetes-risk-prediction/MLP/keep/anchor/anchor_explain_'+str(k)+'.csv'
    keep_temp.to_csv(filename,index=False)
    mean_length = count / 156
print("mean_length:",mean_length)
print("maxlen:",maxlen)
```
