```python
#特征两两相关性分析
df.corr()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True,fmt='.1f',square=True,cmap="YlGnBu")
plt.show()
```

```python
sns.pairplot(df,vars=["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean"])
```

```python
#绘制前十列之间两两的散点图
sns.pairplot(df,hue='diagnosis',vars=["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean"])
```

```python
#单个特征统计分布分析
sns.distplot(df['concave points_mean'])
plt.show()
```

```python
#箱形图
sns.boxplot(x=df['diagnosis'],y=df['area_worst'])
plt.show()
```

```python
#散点图
plt.scatter(x=df["area_worst"],y=df["concave points_worst"])
plt.xlabel('area_worst')
plt.ylabel('concave points_worst')
plt.show()
```

```python
#绘制饼图：半径为0.5，数值保留1位小数
import numpy as np
#类标整数化
count=[212,357]
pie_labels = np.array(['malignant', 'benign'])
plt.pie(count,radius=1.0,labels=pie_labels,autopct='%3.1f%%')
plt.show()
```
