```python
import pandas as pd

data = pd.read_csv('dataset/breast-cancer-wisconsin/test_case/unexplainable_sample_mlp_lower.csv')
data.iloc[:, 1:] = data.iloc[:, 1:].apply(lambda col: col.round(14))
duplicates = data.duplicated()
data[duplicates]
```

```python
cleaned_data = data.drop_duplicates(keep='first')
cleaned_data.to_csv('dataset/breast-cancer-wisconsin/test_case/unexplainable_sample_mlp_lower_cleaned.csv', index=False)
```
