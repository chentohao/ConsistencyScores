```python
import lore

from prepare_dataset import *
from neighbor_generator import *

from sklearn.model_selection import train_test_split

warnings.filterwarnings("ignore")
```

```python
path_data = '../../dataset/early-stage-diabetes-risk-prediction/'

dataset_name = 'diabetes.csv'
dataset = prepare_diabetes_dataset(dataset_name, path_data)

X, y = dataset['X'], dataset['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.preprocessing import StandardScaler
transform = StandardScaler()
X_train[:, 0] = transform.fit_transform(X_train[:, 0].reshape(-1, 1)).ravel()
X_test[:, 0] = transform.transform(X_test[:, 0].reshape(-1, 1)).ravel()

import joblib
blackbox = joblib.load('../../saved_model/early-stage-diabetes-risk-prediction/RF.pkl')

X2E = X_test
y2E = blackbox.predict(X2E)
y2E = np.asarray([dataset['possible_outcomes'][i] for i in y2E])
```

```python
explanations = [] 

# First loop over idx_record2explain, generate explanation, and store the results in explanations
for idx_record2explain in range(156):
    explanation, infos = lore.explain(idx_record2explain, X2E, dataset, blackbox,
                                      ng_function=genetic_neighborhood,
                                      discrete_use_probabilities=True,
                                      continuous_function_estimation=False,
                                      returns_infos=True,
                                      path=path_data, sep=';', log=False)
    explanations.append((explanation, infos))
```

```python
# Then loop over k, and for each k iterate through all explanations
for k in range(1, 32):
    test_data = pd.read_csv('../../dataset/early-stage-diabetes-risk-prediction/test_data.csv')
    count = 0
    maxlen = 0 
    for idx_record2explain in range(156):
        explanation, infos = explanations[idx_record2explain] # Use the previously stored explanation
        dfX2E = build_df2explain(blackbox, X2E, dataset).to_dict('records')
        dfx = dfX2E[idx_record2explain]
        # x = build_df2explain(blackbox, X2E[idx_record2explain].reshape(1, -1), dataset).to_dict('records')[0]
        keys_view = explanation[0][1].keys()
        keys_list = list(keys_view)
        length = len(keys_list)
        count = count + length
        if length > maxlen:
            maxlen = length
        if length > k:
            for j in range(0,k):
                if keys_list[j]=='Age':
                    test_data.at[idx_record2explain,keys_list[j]]=94
                else:
                    test_data.at[idx_record2explain,keys_list[j]]=0
        elif length <= k:
            for j in range(0,length):
                if keys_list[j]=='Age':
                    test_data.at[idx_record2explain,keys_list[j]]=94
                else:
                    test_data.at[idx_record2explain,keys_list[j]]=0
    filename = '../../explain_set/early-stage-diabetes-risk-prediction/RF/lore/lore_explain_'+str(k)+'.csv'
    test_data.to_csv(filename,index=False)
    mean_length = count / 156
print("mean_length:",mean_length)
print("maxlen:",maxlen)
```

```python
# Then loop over k, and for each k iterate through all explanations
for k in range(1, 32):
    test_data = pd.read_csv('../../dataset/early-stage-diabetes-risk-prediction/test_data.csv')
    keep_temp = pd.read_csv('../../dataset/early-stage-diabetes-risk-prediction/keep_temp.csv')
    count = 0
    maxlen = 0
    for idx_record2explain in range(156):
        explanation, infos = explanations[idx_record2explain] # Use the previously stored explanation
        dfX2E = build_df2explain(blackbox, X2E, dataset).to_dict('records')
        dfx = dfX2E[idx_record2explain]
        # x = build_df2explain(blackbox, X2E[idx_record2explain].reshape(1, -1), dataset).to_dict('records')[0]
        keys_view = explanation[0][1].keys()
        keys_list = list(keys_view)
        length = len(keys_list)
        count = count + length
        if length > maxlen:
            maxlen = length
        if length > k:
            for j in range(0,k):
                keep_temp.at[idx_record2explain,keys_list[j]]=test_data.at[idx_record2explain,keys_list[j]]
        elif length <= k:
            for j in range(0,length):
                keep_temp.at[idx_record2explain,keys_list[j]]=test_data.at[idx_record2explain,keys_list[j]]
    filename = '../../explain_set/early-stage-diabetes-risk-prediction/RF/keep/lore/lore_explain_'+str(k)+'.csv'
    keep_temp.to_csv(filename,index=False)
    mean_length = count / 156
print("mean_length:",mean_length)
print("maxlen:",maxlen)
```
